```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link='https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
page
```

```python
soup = bs(page.content,'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
names
```

```python
#Extracting customer names
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())

```

```python
cust_name
```

```python
#Extracting Review title.
title = soup.find_all('a',class_='review-title-content')
for i in range(0,len(title)):
 title.append(title[i].get_text())
```

```python
title
```

```python
# Extracting review ratings
ratings = soup.find_all('i', class_='review-rating')
review_rating = []
for rating in ratings:
    rating_text = rating.find('span', class_='a-icon-alt').get_text()
    review_rating.append(rating_text.split()[0])
    
```

```python
ratings
```

```python
# Extracting review dates
review_dates = soup.find_all('span', class_='review-date')
review_date = [date.get_text() for date in review_dates]
```

```python
review_date
```

```python
df = pd.DataFrame({
    'Customer Name': cust_name,
    #'Review Title': title,
    #'Review Rating': review_rating
})
```

```python
df
```

```python
#dataframe is converted to csv file
df.to_csv(r'G:\reviews.csv',index=True)
```
